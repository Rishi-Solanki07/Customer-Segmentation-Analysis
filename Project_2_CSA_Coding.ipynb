**in top_speed, warranty, and current price are naturally High so won't change that, we will  solve vehile_weight outlier, rest columns has no outliers**

```python
#let's solve vehicle weight outlier with iqr method

q1 = df['vehicle_weight'].quantile(0.25)
q3 = df['vehicle_weight'].quantile(0.75)

iqr = q3-q1

upper_bound = q3 + 1.5*iqr
lower_bound = q1 - 1.5*iqr

df['vehicle_weight'] = np.where(df['vehicle_weight'] > upper_bound, upper_bound, np.where(df['vehicle_weight'] < lower_bound, lower_bound, df['vehicle_weight']))
```

```python
#let's convert our numericsl columns float64, int64 into float32  if this works, that means our numerical columns are clear +. no bad data or anomalies

numerical_columns = df.select_dtypes(include=['int64','float64']).columns
for col in numerical_columns:
  df[col] = df[col].astype('float32')

#let's check catrgorical columns with same way with loop will change the dtype of categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
  df[column] = df[column].astype('category')
```

```python
# Define age groups
age_labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '66+']

# Create a function to map age values to age groups
def map_age_to_group(age):
    if age <= 25:
        return '18-25'
    elif age <= 35:
        return '26-35'
    elif age <= 45:
        return '36-45'
    elif age <= 55:
        return '46-55'
    elif age <= 65:
        return '56-65'
    else:
        return '66+'

# Apply the mapping function to the 'customer_age' column to create a new 'age_group' column
df['age_group'] = df['customer_age'].apply(map_age_to_group)

# Filter data for each vehicle
vehicles = df['vehicle_name'].unique()
for vehicle in vehicles:
    plt.figure(figsize=(8, 6))
    plt.title(f'Service Ratings for {vehicle}')
    plt.xlabel('Age Group')
    plt.ylabel('Service Ratings')
    plt.xticks(ticks=np.arange(len(age_labels)), labels=age_labels, rotation=45, ha='right')

    # Filter data for the current vehicle
    vehicle_data = df[df['vehicle_name'] == vehicle]

    # Iterate over each rating (1 to 5) and plot the corresponding service ratings
    for rating in range(1, 6):
        rating_data = vehicle_data[vehicle_data['service_ratings'] == rating]
        x = [age_labels.index(age_group) for age_group in rating_data['age_group'] if age_group in age_labels]
        y = [rating] * len(x)
        plt.scatter(x, y, label=f'Rating {rating}', alpha=0.7)

    plt.legend()
    plt.tight_layout()
    plt.show()
```

**as we can see we have multicollinearity, so for that we will use cumulative_varience method to solve this problem**  

```python
from sklearn.decomposition import PCA
# Initialize PCA with all components
pca = PCA()

# Fit PCA to the data
pca.fit(dfa)

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(cumulative_variance, marker='o', linestyle='-', color='b')
plt.title('Cumulative Explained Variance Plot')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

**as we can see the variance stops on 6th component so we will use only first 6 components in k_means clustering and to reduce the rest components we will use PCA method**

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Number of components to retain
n_components = 6

# Initialize PCA with desired number of components
pca = PCA(n_components=n_components)

# Fit PCA to the data and transform the data
pca_features = pca.fit_transform(dfa)
```

```python
print(pca_features)
```

**as we can see it is not dataframe let's convert this into a dataframe and we will use that data-frame into next step**

```python
# Convert the numpy array to a DataFrame with column names pca-1, pca-2, ..., pca-6
pca_df = pd.DataFrame(data=pca_features, columns=[f"pca-{i}" for i in range(1, n_components + 1)])

# Print the DataFrame
pca_df.head(7)
```

**now we will decide how many customer grouping we are going to do, and for that we will use elbow curve plot**

```python
from sklearn.cluster import KMeans

# Range of clusters to try
k_values = range(1, 11)  # Try clusters from 1 to 10

# Initialize list to store inertia values
inertia_values = []

# Calculate inertia for each value of K
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_df)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(k_values, inertia_values, marker='o')
plt.title('Elbow Curve for KMeans Clustering')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

**as per the result of elbow curve we will do 5 grouping this time**

```python
from sklearn.cluster import KMeans

# Number of clusters
n_clusters = 5

# Initialize KMeans with desired number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Fit KMeans to the PCA-transformed data
kmeans.fit(pca_df)

# Assign clusters to each data point
clusters = kmeans.predict(pca_df)

# Add the cluster labels to the original dataframe
pca_df['cluster'] = clusters

# Explanation:
# 1. We specify the number of clusters we want to create, in this case, 5 clusters.
# 2. We initialize the KMeans algorithm with the specified number of clusters.
# 3. We fit the KMeans model to the PCA-transformed data to identify the cluster centers.
# 4. We predict the cluster labels for each data point.
# 5. We add the cluster labels as a new column in the dataframe.
# 6. Finally, we display the dataframe with the cluster labels assigned to each data point.
```

**let's plot the grouping result**

```python
# Plotting the clusters
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['pca-1'], pca_df['pca-2'], c=pca_df['cluster'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('KMeans Clustering')
plt.colorbar(label='Cluster')
plt.show()
```
